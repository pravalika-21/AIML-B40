```python
import numpy as np
array1=np.array([10,20,30])
#sine function
print("sine:.",np.log(array1))
#logarithm
print("natural logarithm: ",np.log(array1))
print("base-10 logarithm: ",np.log10(array1))
print("base-2 logarithm: ",np.log2(array1))
#exponential
print("exponential: ",np.exp(array1))
```

```python
name=['kiran','arun','vijay','varun']
age=[25,30,40,45]
height =[145, 151, 165, 173]
weight=[45, 55, 65, 75]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(weight,height)
plt.title("Relationship between weight and height of the patients")
plt.ylabel("height in meters")
plt.xlabel("height of patients")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,4))
plt.title("name, weight in Kgs",fontsize=16,fontstyle='italic')
plt.bar(x=name,height=weight,width=0.6,color='darkred',edgecolor='k',alpha=0.6)
plt.xlabel("name",fontsize=15)
plt.ylabel("weight(in kgs)",fontsize=15)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.show()
```

```python
import numpy as np

plt.figure(figsize=(7,5))

# Main plot function 'hist'

plt.hist(weight, color='red', edgecolor='k')

plt.title("Histogram of patient weight", fontsize=18)

plt.xlabel("Weight in kgs", fontsize=15)

plt.xticks(fontsize=15)

plt.yticks(fontsize=15)

plt.show()
```
